<h1> libraries for plots </h1>

```python
import matplotlib.pyplot as plt
import numpy as np
```

<h1> relation between 2 joints in a 2d plane</h1>

```python

# Define position vectors
P0 = np.array([1, 2])  # Example point
D1 = np.array([3, 4])  # Direction vector
P1 = P0 + D1  # Compute P1 as P0 + D1

# Rotation function
def rotate_point(point, origin, angle):
    angle_rad = np.radians(angle)
    rotation_matrix = np.array([[np.cos(angle_rad), -np.sin(angle_rad)],
                                [np.sin(angle_rad), np.cos(angle_rad)]])
    return origin + np.dot(rotation_matrix, (point - origin))

# Rotate P1 and D1 by a given angle
angle = -45  # Rotation angle in degrees
P1_rotated = rotate_point(P1, P0, angle)
D1_rotated = P1_rotated - P0

# Plot the points
fig, ax = plt.subplots()
ax.scatter([P0[0], P1[0], P1_rotated[0]], [P0[1], P1[1], P1_rotated[1]], color=['r', 'b', 'purple'], label=['P0', 'P1', 'P1 Rotated'])
ax.quiver(P0[0], P0[1], D1[0], D1[1], angles='xy', scale_units='xy', scale=1, color='g', label='D1')
ax.quiver(P0[0], P0[1], D1_rotated[0], D1_rotated[1], angles='xy', scale_units='xy', scale=1, color='orange', label='D1 Rotated')

# Set limits and labels
ax.set_xlim(0, max(P0[0], P1[0], P1_rotated[0]) + 2)
ax.set_ylim(0, max(P0[1], P1[1], P1_rotated[1]) + 2)
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_title('Rotation of P1 and D1 around P0')
ax.legend()
ax.grid()

plt.show()
print(np.sqrt(D1[0]**2 + D1[1]**2))
print(np.sqrt(D1_rotated[0]**2 + D1_rotated[1]**2))
print("Linear rotation")
```
